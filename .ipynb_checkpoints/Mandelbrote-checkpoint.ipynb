```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, r):
    return r*x*(1 - x)
```

```python
plt.figure(figsize=(10,6))
rates = list(np.linspace(2.8,4.0,5000))
population = []
for rate in rates:
    y = [0.5]
    for i in range(30):
        y.append(f(y[i], rate))
    y = list(set(y[15:]))
    x = [rate]*len(y)
    plt.scatter(x, y, marker='o', s=0.1)
```

```python
N=1000
y=[]
y.append(0.5)
for r in np.arange(2.8,4.0,0.001):  
    for n in np.arange(1,N,1):       
        y.append(round(r*y[n-1]*(1-y[n-1]),4)) 
    y=y[N-250:N]
    x=[r]*250
    plt.plot(x,y, color='black', linestyle=' ', marker='.', markersize=1) 
plt.title("   2,8<= r<=4,0,0<=x<=1")
plt.xlabel("r")
plt.ylabel("y")
plt.axvline(x=3.01,color='black',linestyle='--')
plt.axvline(x=3.45, color='black',linestyle='--')
plt.axvline(x=3.6,color='black',linestyle='--')
plt.axvline(x=3.7,color='black',linestyle='--')
plt.axvline(x=3.8,color='black',linestyle='--')
plt.axvline(x=3.9,color='black',linestyle='--')
plt.show()
```
