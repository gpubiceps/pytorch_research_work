```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
from scipy.integrate import simps
import pandas as pd
from scipy import signal
```

```python
def create_ECG(filename_ecg):
    ECG = scipy.io.loadmat(filename_ecg) #загрузка .mat-файла
    gain = 200
    signal_II = ECG['val'][0,:]/gain
    return signal_II

Signal = create_ECG('116m.mat')
```

```python
def create_marks(filename_marks):
    data = pd.read_csv(filename_marks, sep="\t")
    Marks = []
    for i in range(len(data)):
        mark = str(data.iloc[i])
        Marks.append(int(mark.split()[8:9][0]))
    return Marks

Marks = create_marks('annotations.txt')
```

```python
def create_class(filename_class):
    data = pd.read_csv(filename_class, sep="\t")
    Class = []
    for i in range(len(data)):
        Class_i = str(data.iloc[i])
        Class.append(Class_i.split()[9:10][0])
    return Class

Class = create_class('annotations.txt')
```

```python
def duration_QRS(Signal, Marks, Class, fs = 250, lp_order = 6, lp_cutoff = 30, hp_order = 6, hp_cutoff = 2, lp_x2 = False,
                 k1 = 2.5, k2 = 7.2, n_qrs = [], len_qrs = 100, degree = 2, poly_len = 30, polyshift = 0.0, plot = False):
#***********************************************************************************    
    def butter_lowpass(lp_cutoff, fs, lp_order):
        nyq = 0.5 * fs
        normal_cutoff = lp_cutoff / nyq
        b, a = signal.butter(lp_order, normal_cutoff, btype='low', analog=False)
        return b, a

    def butter_lowpass_filter(data, lp_cutoff, fs, lp_order):
        b, a = butter_lowpass(lp_cutoff, fs, lp_order)
        y = signal.lfilter(b, a, data)
        return y
    
    b, a = butter_lowpass(lp_cutoff, fs, lp_order)
    if plot == True:
        w, h = signal.freqz(b, a, worN=8000)
        plt.figure(figsize=(11,3))
        plt.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
        plt.plot(lp_cutoff, 0.5*np.sqrt(2), 'ko')
        plt.axvline(lp_cutoff, color='k')
        plt.xlim(0, 0.5*fs)
        plt.title("Lowpass Filter Frequency Response")
        plt.xlabel('Frequency [Hz]')
        plt.grid()

    T = 4.0
    n = int(T * fs)
    t = np.linspace(0, T, n, endpoint=False)

    signal_lowpass = butter_lowpass_filter(Signal, lp_cutoff, fs, lp_order)
    
    if lp_x2 == True:
        signal_lowpass_invert = signal_lowpass[::-1]
        signal_lowpass_x2 = butter_lowpass_filter(signal_lowpass_invert, lp_cutoff, fs, lp_order)[::-1]
    if plot == True:
        plt.figure(figsize=(11,3))
        plt.plot(t[500:950], Signal[500:950], 'b-', linewidth=1, label='data')
        plt.plot(t[500:950], signal_lowpass[500:950], 'r-', linewidth=1, label='filtered data')
        plt.xlabel('Time [sec]')
        plt.grid()
        plt.legend()
        plt.show()
    
    if lp_x2 == True and plot == True:
        plt.figure(figsize=(11,3))
        plt.plot(t[500:950], Signal[500:950], 'b-', linewidth=1, label='data')
        plt.plot(t[500:950], signal_lowpass_x2[500:950], 'r-', linewidth=1, label='double filtered data')
        plt.xlabel('Time [sec]')
        plt.grid()
        plt.legend()
        plt.show()
#***********************************************************************************
    def butter_highpass(hp_cutoff, fs, hp_order):
        nyq = 0.5 * fs
        normal_cutoff = hp_cutoff / nyq
        b, a = signal.butter(hp_order, normal_cutoff, btype='high', analog=False)
        return b, a

    def butter_highpass_filter(data, hp_cutoff, fs, hp_order):
        b, a = butter_highpass(hp_cutoff, fs, hp_order)
        y = signal.lfilter(b, a, data)
        return y
    
    b, a = butter_highpass(hp_cutoff, fs, hp_order)
    if plot == True:
        w, h = signal.freqz(b, a, worN=8000)
        plt.figure(figsize=(11,3))
        plt.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
        plt.plot(hp_cutoff, 0.5*np.sqrt(2), 'ko')
        plt.axvline(hp_cutoff, color='k')
        plt.xlim(0, hp_cutoff + 10)
        plt.title("Highpass Filter Frequency Response")
        plt.xlabel('Frequency [Hz]')
        plt.grid()
    plt.figure()
    plt.plot(a,label=a)
    plt.plot(b,label=b)
    plt.legend()
    T = 4.0
    n = int(T * fs)
    t = np.linspace(0, T, n, endpoint=False)
    
    if lp_x2 == False:
        signal_highpass = butter_highpass_filter(signal_lowpass, hp_cutoff, fs, hp_order)
    else:
        signal_highpass = butter_highpass_filter(signal_lowpass_x2, hp_cutoff, fs, hp_order)
    #Вторая фильрация в обратную сторону    
    signal_highpass_invert = signal_highpass[::-1]
    signal_highpass_x2 = butter_highpass_filter(signal_highpass_invert, hp_cutoff, fs, hp_order)[::-1]
    
    if plot == True:
        plt.figure(figsize=(11,3))
        plt.plot(t[500:950], Signal[500:950], 'b-', linewidth=1, label='data')
        plt.plot(t[500:950], signal_highpass_x2[500:950], 'r-', linewidth=1, label='filtered data')
        plt.xlabel('Time [sec]')
        plt.grid()
        plt.legend()
        plt.show()
#***********************************************************************************
    #Первая производная
    diff_I = []

    #diff_I.append(0.0)
    #diff_I.append(0.0)
    if lp_x2 == False:
        for i in range(2,len(signal_lowpass)):
            diff_I.append(signal_lowpass[i]-signal_lowpass[i-2])
    else:
        for i in range(2,len(signal_lowpass_x2)):
            diff_I.append(signal_lowpass_x2[i]-signal_lowpass_x2[i-2])
    diff_I.append(0.0) #Добавим два нуля в конец 
    diff_I.append(0.0)

    #Вторая производная
    diff_II = []
    #diff_II.append(0.0)
    #diff_II.append(0.0)
    for i in range(2,len(diff_I)):
        diff_II.append(diff_I[i]-diff_I[i-2])
    diff_II.append(0.0)
    diff_II.append(0.0)
    
    diff_I_abs = np.abs(diff_I) * k1
    diff_II_abs = np.abs(diff_II) * k2
    
    #Сумма модулей сигнала и двух производных
    sum_qrs = np.abs(signal_highpass_x2) + diff_I_abs + diff_II_abs
    
#***********************************************************************************
    #Массив с QRS комплексами
    def create_QRS_array(QRS_len = 100):
        QRS = []
        for i in Marks[1:]:
            QRS_i = []
            QRS_i = list(sum_qrs[i-QRS_len : i+QRS_len])
            QRS.append(QRS_i)
        return QRS

    QRS = create_QRS_array(QRS_len = len_qrs)
    
    coord = [] #Координаты х начала и конца комплексов
    for i in n_qrs:
        coord.append([Marks[i + 1] - len_qrs, Marks[i + 1] + len_qrs])
#***********************************************************************************    
    if plot == True:
        plt.figure(figsize=(9,5))
        plt.plot(Signal[coord[0][0] : coord[0][1]], linewidth=2, label='signal')
        plt.plot(signal_highpass_x2[coord[0][0] : coord[0][1]]+6, linewidth=2, label='signal_highpass_x2')
        plt.plot(np.abs(signal_highpass_x2[coord[0][0] : coord[0][1]])+8, linewidth=2, label='signal_highpass_abs_x2')
        plt.plot(diff_I_abs[coord[0][0] : coord[0][1]]+11, linewidth=2, label='diff_I_abs')
        plt.plot(diff_II_abs[coord[0][0] : coord[0][1]]+14, linewidth=2, label='diff_II_abs')
        plt.grid()
        plt.legend()
    
    #Полином n степени
    def polynom(data, len_qrs, poly_len, degree = 2):
        k = np.array([i for i in range(len_qrs - poly_len,len_qrs + poly_len)])
        z = np.polyfit(k, data, degree)
        polinom = np.poly1d(z)
        t = np.array([i for i in range(0,2*len_qrs)])
        polinom_y = polinom(t)
        return polinom_y
    
    #Пересечение с нулем, используя максимум
    def across_nul(data, len_qrs, poly_len):
        across = []
        maximum_y = np.amax(data[len_qrs - poly_len:len_qrs + poly_len])
        maximum_x = 0
        count = 0
        for i in data[len_qrs - poly_len:len_qrs + poly_len]:
            if i == maximum_y:
                maximum_x = count
                break
            count += 1
        maximum_x = maximum_x + len_qrs - poly_len
        for i in range(maximum_x,0,-1):
            if data[i] < 0.0:
                across.append(i)
                break
        for i in range(maximum_x,len(data),1):
            if data[i] < 0.0:
                across.append(i)
                break
        return across
    
    N = len(n_qrs)
    if N%2 != 0:
        N = int(N/2+1)
    else:
        N = int(N/2)
    plt.figure(figsize=(7*N,7*N))
    for i in range(len(n_qrs)):
        poly = polynom(QRS[n_qrs[i]][len_qrs - poly_len:len_qrs + poly_len], len_qrs, poly_len, degree)
        nuli = across_nul(poly-polyshift, len_qrs, poly_len)
        if len(nuli) == 2:
            t_poly = np.array([i for i in range(nuli[0],nuli[1])])
        else:
            t_poly = np.array([i for i in range(len_qrs - poly_len,len_qrs + poly_len)])

        plt.subplot(N,2,i+1)
        plt.plot(Signal[coord[i][0] : coord[i][1]], linewidth=2, label='signal')
        plt.plot(QRS[n_qrs[i]], 'k', linewidth=2, label='sum_qrs')
        if len(nuli) == 2:
            plt.plot(t_poly, poly[nuli[0]:nuli[1]]-polyshift, '--', linewidth=2, label='polynomial')
        else:
            plt.plot(t_poly, poly[len_qrs - poly_len:len_qrs + poly_len]-polyshift, '--', linewidth=2, label='polynomial')
        plt.axvline(len_qrs - poly_len, color='k')
        plt.axvline(len_qrs + poly_len, color='k')
        if len(nuli) == 2:
            plt.scatter(nuli, [0,0], color='r', s=40, marker='o')
            plt.text(0, 2, 'Duration = '+str((nuli[1] - nuli[0])/250))
        if len(nuli) == 1:
            plt.scatter(nuli, [0], color='r', s=40, marker='o')
        plt.grid(True)
        plt.legend()
        plt.text(0, 4, str(Class[n_qrs[i]+1]), fontsize=14)
        plt.text(0, 5, str(n_qrs[i]), fontsize=14)
```

```python
duration_QRS(Signal, Marks, Class, fs = 250, lp_order = 2, lp_cutoff = 30, hp_order = 2, hp_cutoff = 2, lp_x2 = True,
                 k1 = 1.9, k2 = 4.2, n_qrs = [127,130,549,550,551,381], len_qrs = 100, degree = 2, poly_len = 25,
                   polyshift = 0.0, plot = True)
```
